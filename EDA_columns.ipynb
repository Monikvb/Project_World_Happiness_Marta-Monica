```python
# Normalizar los nombres de las columnas para happy_2018 y happy_2019
for df in [happy_2018, happy_2019]:
    df.rename(columns={'Country or region': 'Country'}, inplace=True)

# Normalizar los nombres de las columnas para happy_2022
happy_2022.rename(columns={'Happiness score': 'Happiness Score', 
                           'Explained by: GDP per capita': 'GDP per capita', 
                           'Explained by: Social support': 'Social support',
                            'Explained by: Healthy life expectancy': 'Healthy life expectancy',
                            'Explained by: Freedom to make life choices': 'Freedom to make life choices',
                            'Explained by: Generosity': 'Generosity', 
                            'Explained by: Perceptions of corruption': 'Perceptions of corruption'
                        }, inplace=True)

# Lista de DataFrames
dataframes = [happy_2015, happy_2016, happy_2017, happy_2018, happy_2019, happy_2020, happy_2021, happy_2022, happy_2023]

# Columnas que queremos mantener
common_columns = [
    'Country', 'Happiness Score', 'GDP per capita', 'Social support', 'Healthy life expectancy',
    'Freedom to make life choices', 'Perceptions of corruption', 'Generosity', 'Año'
]
# Seleccionar solo las columnas comunes y concatenar todos los DataFrames
filtered_dataframes = [df[common_columns] for df in dataframes]
combined_df = pd.concat(filtered_dataframes, ignore_index=True)

# Guardar el DataFrame combinado en un único CSV
combined_df.to_csv("combined_happiness.csv", index=False)

# Imprimir las primeras filas del DataFrame resultante para verificar
print(combined_df.head())
```
